```python
result_directory = '/content/drive/MyDrive/NLP Projects/Sentiment analysis/results/'
```

```python
import torch
from transformers import AutoModelForSequenceClassification, BertTokenizer

tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

model = AutoModelForSequenceClassification.from_pretrained(result_directory)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

def predict(input_text):
  inputs = tokenizer(input_text, return_tensors="pt").to(device)

  # انجام پیش‌بینی
  with torch.no_grad():
      outputs = model(**inputs)
      logits = outputs.logits
      predictions = torch.argmax(logits, dim=-1)

  return predictions
```

Positive Texts

```python
predict("This movie was amazing! I really enjoyed watching it.")
```

```python
predict("The products from this brand are of very high quality, and I'm always satisfied with them.")
```

```python
predict("The customer service at this store is outstanding. I always feel good when I go there.")
```

Negative Texts

```python
predict("This book was very boring, and I couldn't finish it.")
```

```python
predict("The quality of this product is very low, and I regret my purchase.")
```

```python
predict("My experience at this restaurant was not good at all. The food was cold and tasteless.")
```
